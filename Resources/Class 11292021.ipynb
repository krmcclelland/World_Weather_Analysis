```python
from requests import get

url = 'https://api.datamuse.com/words?max=1000&rel_rhy='

word = 'shoe'

get_url = url + word

print("making a GET request to:", get_url)

rhymes = get(get_url).json()

print(rhymes)

for rhyme in rhymes:
    print(rhyme)
    print('=====')
```

```python
# or as a function!

def getRhymes(word):
    rhymes = get('https://api.datamuse.com/words?max=1000&rel_rhy=' + word).json()
    return rhymes

rhymes = getRhymes('dog')

print(rhymes)
```
